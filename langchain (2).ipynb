```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline
from langchain_core.prompts import PromptTemplate

# FIXED pipeline
pipe = pipeline(
    "text-generation",
    model="gpt2",
    max_new_tokens=100   # ✅ FIX HERE
)

llm = HuggingFacePipeline(pipeline=pipe)

# Memory storage
chat_history = []

prompt = PromptTemplate(
    input_variables=["history", "input"],
    template="""
You are a helpful assistant.

Conversation:
{history}

User: {input}
Assistant:"""
)

def chat(user_input):
    history_text = "\n".join(chat_history)

    final_prompt = prompt.format(
        history=history_text,
        input=user_input
    )

    response = llm.invoke(final_prompt)

    # Clean response
    response = response.replace(final_prompt, "").strip()

    chat_history.append(f"User: {user_input}")
    chat_history.append(f"Assistant: {response}")

    return response

# Test
print(chat("Hi, my name is Deeksha"))
print(chat("What is my name?"))
```

```python
from langchain_core.tools import Tool
from langchain.agents import initialize_agent
from langchain.agents import AgentType

# Create a simple tool
def calculator(query):
    return str(eval(query))

tool = Tool(
    name="Calculator",
    func=calculator,
    description="Useful for solving math problems"
)

# Create agent
agent = initialize_agent(
    tools=[tool],
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

# Run
response = agent.run("What is 25 + 10 * 2?")
print(response)
```

```python
# Tool (Calculator)
def calculator(query):
    return str(eval(query))

# Agent logic
def agent(query):
    if any(op in query for op in ["+", "-", "*", "/"]):
        return f"Using Calculator Tool: {calculator(query)}"
    else:
        return llm.invoke(query)

# Test
print(agent("25 + 10 * 2"))
print(agent("Explain Artificial Intelligence"))
```
